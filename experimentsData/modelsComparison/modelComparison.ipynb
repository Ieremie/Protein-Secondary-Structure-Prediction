```python
import pickle 
import os

cwd = os.getcwd()  # Get the current working directory (cwd)
files = os.listdir(cwd)  # Get all the files in that directory
print("Files in %r: %s" % (cwd, files))
```

```python
m1H = pickle.load( open( "mLayer1", "rb" ) )
m3H = pickle.load( open( "mLayer3", "rb" ) )
m5H = pickle.load( open( "mLayer5", "rb" ) )
m7H = pickle.load( open( "mLayer7", "rb" ) )
m9H = pickle.load( open( "mLayer9", "rb" ) )
m11H = pickle.load( open( "mLayer11", "rb" ) )
```

```python
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.plot(m1H['accuracy'])
plt.plot(m1H['val_accuracy'])

plt.plot(m3H['accuracy'])
plt.plot(m3H['val_accuracy'])

plt.plot(m5H['accuracy'])
plt.plot(m5H['val_accuracy'])

plt.plot(m7H['accuracy'])
plt.plot(m7H['val_accuracy'])

plt.plot(m9H['accuracy'])
plt.plot(m9H['val_accuracy'])

plt.plot(m11H['accuracy'])
plt.plot(m11H['val_accuracy'])

plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
# Plot training & validation loss values

plt.plot(m1H['loss'])
plt.plot(m1H['val_loss'])

plt.plot(m3H['loss'])
plt.plot(m3H['val_loss'])

plt.plot(m5H['loss'])
plt.plot(m5H['val_loss'])

plt.plot(m7H['loss'])
plt.plot(m7H['val_loss'])

plt.plot(m9H['loss'])
plt.plot(m9H['val_loss'])

plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
#plt.savefig(lossName)
plt.show()

```
